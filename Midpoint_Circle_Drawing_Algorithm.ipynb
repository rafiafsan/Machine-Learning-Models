```python
import matplotlib.pyplot as plt
```

```python
def midpoint_circle(x_center,y_center,radius):
  x=0
  y=radius
  d= 1- radius

  while y > x:
      if d < 0:
       d+=3 * x+3
      else:
       d+=3 * (x-y) + 5
       y-=1
      x+=1

      plt.plot(x_center + x, y_center + y, 'ro')
      plt.plot(x_center - x, y_center + y, 'ro')
      plt.plot(x_center + x, y_center - y, 'ro')
      plt.plot(x_center - x, y_center - y, 'ro')
      plt.plot(x_center + y, y_center + x, 'ro')
      plt.plot(x_center - y, y_center + x, 'ro')
      plt.plot(x_center + y, y_center - x, 'ro')
      plt.plot(x_center - y, y_center - x, 'ro')


x_center = int(input("Enter x coordiates of center: "))
y_center = int(input("Enter y coordiates of center: "))
radius = int(input("Enter Radius:"))
midpoint_circle(x_center,y_center,radius)

plt.axis('equal')
plt.title(f'Midpoint Circle (Center: ({x_center}, {y_center}), Radius: {radius})')
plt.grid(True)
plt.show()
```
